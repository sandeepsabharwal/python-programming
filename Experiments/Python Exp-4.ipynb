## 1. Find a factorial of a number

```python
num = int(input("Enter a number"))
fact = 1
for i in range(1, num + 1):
    fact *= i
print(fact)
```

## 2. Find whether the given number is Armstrong number

```python
num = int(input("Enter a number: "))
order = len(str(num))
sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum += digit ** order
    temp //= 10
if num == sum:
    print("Armstrong number")
else:
    print("Not Armstrong number")
```

## 3. Print Fibonacci series up to given term

```python
n = int(input("enter the number of terms"))
a = 0
b = 1
for i in range(n):
    print(a, " ")
    a, b = b, a+b
```

## 4. Write a program to find if the given number is prime number or not

```python
num = int(input("Enter a number: "))
if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

## 5. Check whether given number is palindorme or not

```python
n = int(input("Enter a number: "))
reverse = 0
temp = n
while temp != 0:
    reverse = (reverse * 10) + (temp % 10)
    temp = temp // 10
if n == reverse:
    print("Paindrome")
else:
    print("Not Palindrome")
```

## 6. Write a program to print the sum of digits 

```python
n = int(input("Enter a number: "))
sum = 0
s = str(n)
temp = n
for i in s:
    digit = temp % 10
    temp = temp // 10
    sum += digit
print(sum)
```

## 7. Count an print all numbers divisible by 5 and 7 between 1 and 100

## 9. Print all prime number between 1 to 100

```python
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
