1. Write a Python function to find the maximum and minimum numbers from a 
sequence of numbers.  (Note: Do not use built-in functions.)  

```python
def find_max_min(numbers):
    if len(numbers) == 0:
        return None
    
    max_num = numbers[0]
    min_num = numbers[0]
    
    for num in numbers:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num
    
    return max_num, min_num


nums = [10, 6, 8, 90, 12, 56]
print(find_max_min(nums))
```

2. Write a Python function that takes a positive integer and returns the sum of the 
cube of all the positive integers smaller than the specified number.  

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total



print(sum_of_cubes(5))
```

3. Write a Python function to print 1 to n using recursion. (Note: Do not use loop) 

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)



print_numbers(5)
```

4. Write a recursive function to print Fibonacci series upto n terms.

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


def print_fibonacci(n):
    for i in range(n):
        print(fibonacci(i), end=" ")



print_fibonacci(7)
```

5. Write a lambda function to find volume of cone.

```python
import math

volume_cone = lambda r, h: (1/3) * math.pi * r**2 * h


print(volume_cone(3, 5))
```

6. Write a lambda function which gives tuple of max and min from a list.  
Sample input: [10, 6, 8, 90, 12, 56]  
Sample output: (90,6) 

```python
max_min = lambda lst: (
    max(lst), min(lst)
)

# Sample input
data = [10, 6, 8, 90, 12, 56]
print(max_min(data))
```
