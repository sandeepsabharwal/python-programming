## 1. Check whether a given number is divisible by 3 and 5 both

```python
a = int(input("Enter a number"))
if a % 5 == 0 and a % 3 == 0:
    print("Divisible")
else:
    print("Non-Divisible")
```

## 3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.

```python
a = int(input("Enter a number"))
b = int(input("Enter a number"))
if (a > b):
    print(a, "is greater")
elif (b > a):
    print(b, "is greater")
else:
    print(a, "and", b, "are equal")
```

```python
a = int(input("Enter a number"))
b = int(input("Enter a number"))
c = int(input("Enter a number"))
if (a == b) or (b == c) or (a == c):
    print("Two or Three values are same")
else:
    if (a > b) and (a > c):
        print(a, "is greater")
    elif (b > a) and (b > c):
        print(b, "is greater")
    else:
        print(c, "is greater")
```

## 5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots.

```python
import math
a = int(input("Enter a number"))
b = int(input("Enter a number"))
c = int(input("Enter a number"))
D = (b * b) - 4 * a * c
if (D > 0):
    print("Two real and distict roots")
    root1 = (-b + math.sqrt(D)) / (2 * a)
    root2 = (-b - math.sqrt(D)) / (2 * a)
    print(root1, ",", root2, "are the roots")
elif (D == 0):
    print("Real and equal roots")
    root1 = (-b + math.sqrt(D)) / (2 * a)
    print(root1, "is the roots")
else:
    print("Imaginary roots")
    root1 = (-b + math.sqrt(D)) / (2 * a)
    root2 = (-b - math.sqrt(D)) / (2 * a)
    print(root1, ",", root2, "are the roots")
```

## 6. Find whether a given year is a leap year or not.

```python
year = int(input("Enter any year"))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not leap year" )
```

## 7. Write a program which takes any date as input and display next date of the calendar

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days_in_month[1] = 29
day += 1
if day > days_in_month[month - 1]:
    day = 1
    month += 1
    if month > 12:
        month = 1
        year += 1
print("Next date is:", day, "/", month, "/", year)
```
