## 6. Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem

```python
import math
h = int(input("Enter the height of the triangle"))
b = int(input("Enter the base of the triangle"))
sum = (h * h) + (b * b)
c = math.sqrt(sum)
print(c)
```

## 8. Write a program to find the area of triangle when lengths of sides are given

```python
import math
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print(area)
```

## 9. Write a program to convert given seconds into hours, minutes and remaining seconds.

```python
time = int(input("Enter the seconds"))
hour = time // 3600
rem_time = time % 3600
min = rem_time // 60
sec = rem_time % 60
print(hour, "hour", min, "minutes", sec, "seconds")
```

## 12. Write a program to print truth table for bitwise operators (&, |, and ^ operators)

```python
print("A B  A&B  A|B  A^B")
print("-------------------")
for a in range(2):
    for b in range(2):
        and_op = a & b
        or_op  = a | b
        xor_op = a ^ b

        print(a, b, "  ", and_op, "  ", or_op, "  ", xor_op)
```

## 13. Write a program to find the left shift and right shift values of a given number

```python
num = int(input("Enter a number"))
shift = int(input("Enter the number of shifts"))
left_shift = num << shift
right_shift = num >> shift
print(left_shift)
print(right_shift)
```
