## 6. Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem

```python
import math

a = float(input("Enter base: "))
b = float(input("Enter height: "))
c = math.sqrt(a**2 + b**2)
print("Hypotenuse:", c)
```

## 8. Write a program to find the area of triangle when lengths of sides are given

```python
import math

a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of triangle:", area)
```

## 10. Write a program to swap two numbers without takinf additional variable.

```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

x, y = y, x
print("After swapping: x =", x, "y =", y)
```

## 11. Write a program to find the sum of first n natural numbers.

```python
n = int(input("Enter n: "))
total = n * (n + 1) // 2
print("Sum of first", n, "natural numbers:", total)
```

## 12. Write a program to print truth table for bitwise operators (&, |, and ^ operators)

```python
print("x y | x&y x|y x^y")
for x in [0, 1]:
    for y in [0, 1]:
        print(x, y, "|", x & y, " ", x | y, " ", x ^ y)
```

## 13. Write a program to find the left shift and right shift values of a given number

## 14. Using membership operator find whether a given number is in sequence (10,20,56,78,89)

```python
sequence = (10, 20, 56, 78, 89)
num = int(input("Enter a number: "))

if num in sequence:
    print(num, "is in the sequence")
else:
    print(num, "is not in the sequence")
```
