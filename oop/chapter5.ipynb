## EXAM QUESTION

Difference between underscore and double underscore

how would you do add operator function overloading in python



```python
class MyNumber:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        if isinstance(other, int):
            # Step 1: Convert each digit to its corresponding alphabet (0=A, 9=J)
            alphabets = ''.join(chr(int(d) + ord('A')) for d in str(self.value))
            print(f"Step 1 - Number {self.value} → Alphabet String: {alphabets}")

            # Step 2: Convert each alphabet back to its numerical digit
            nums = [ord(c) - ord('A') for c in alphabets]
            print(f"Step 2 - Alphabet '{alphabets}' → Digits: {nums}")

            # Step 3: Reconstruct the number and add 1
            reconstructed = int(''.join(map(str, nums)))
            result_num = reconstructed + other
            print(f"Step 3 - Reconstructed: {reconstructed}, After adding {other}: {result_num}")

            # Step 4: Convert final number's digits back to alphabet string
            final_alpha = ''.join(chr(int(d) + ord('A')) for d in str(result_num))
            print(f"Step 4 - Number {result_num} → Final Alphabet String: {final_alpha}")

            return MyNumber(result_num)

        return NotImplemented

    def __repr__(self):
        return f'MyNumber({self.value})'


# --- Run ---
n = MyNumber(590024428)
print(f"Input: {n}\n")
result = n + 1
print(f"\nFinal Result: {result}")
```

## class method

Class variables are shared by all objects of a class, making them useful when you want a single common value across every instance instead of duplicating it. They save memory, ensure consistency, and make code easier to manage compared to instance variables that change per object.

🔑 What is a Class Variable?
- Defined inside a class but outside methods (like __init__).
- Belongs to the class itself, not to individual objects.
- Shared across all instances of that class.


```python
class Student:
    school = "ABC School"   # Class variable

    def __init__(self, name):
        self.name = name    # Instance variable

s1 = Student("Sandeep")
s2 = Student("Keshav")

print(s1.school)  # ABC School
print(s2.school)  # ABC School
```

```python
class student:
    school_name = "ABC School"                # Class variable
    def __init__(self, name, age):
        self.name = name                      # Instance variable
        self.age = age
s1 = student("Alice", "20")
s2 = student("Bob", "22")     

print(s1.name, s2.name)  # Alice Bob
print(s1.age, s2.age)    # 20 22
```

```python
class Person:
    def __init__(self, name, age_years):
        self.name = name
        self.age_years = age_years

    def convertToDays(self):
        return self.age_years * 365

p1 = Person("Sandeep", 19)
print(f"{p1.name}'s age in days is: {p1.convertToDays()} days")
```

##  Variable Scope ka Matlab
Scope ka matlab hai ek variable kahaan tak accessible hai — matlab program ke kis part mein usko use kar sakte ho.
Python mein scope decide karta hai ki variable "visible" hai ya nahi.
####  Types of Scope (LEGB Rule)
1. Local Scope
- Jo variable function ke andar banaya gaya hai.
- Sirf usi function ke andar use ho sakta hai.
2. Enclosed Scope
- Jab ek function ke andar dusra function ho (nested function).
- Inner function outer function ke variable ko access kar sakta hai.
3. Global Scope
- Jo variable function ke bahar banaya gaya hai.
- Pure program mein accessible hota hai.
4. Built-in Scope
- Python ke khud ke predefined names (jaise len, print, range).
- Ye hamesha available hote hain.
